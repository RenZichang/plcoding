### 2.4 误块率最优性

仔细观察SC译码算法中的条件部分，可以发现并不是我们之前进行Bayes推理时所采用的随机变量$U_{1:i-1}$的真正取值$u_{1:i-1}$，而是判决结果$\hat{u}_{1:i-1}$。这个问题是不可避免的，因为译码器不可能在译码成功前，就准确地知道$u_{1:N}$是多少，否则它就没有必要译码了。那这样自然会带给我们一个疑问：在条件部分填充不准确的估计值，是否会影响译码正确率呢？

为了解释这个问题，我们引入一个概念：Genie-aided译码。它指的是我们通过某种神奇的手段，在译码过程中就使用真正的取值$u_{1:i-1}$，显然这种做法才是我们最开始所期望的做法，并且它应该具有良好的性能。但接下来我们将指出，Genie-aided译码的误块率（Blcok Error Rate，BLER）与SC译码的误块率是相等的。

证明过程十分简单，记Genie-aided译码的第$i$步的判决错误率为$e_i$，那么对所有$N$个随机变量$U_{1:N}$进行判决后，全部正确的概率为：$\prod_{i=1}^N (1-e_i)$。然后再考虑SC译码，若某次译码成功没有发生错误，说明条件部分总是正确的，说明此时与Genie-aided译码每一步所使用的条件概率分布都相同，全部判决正确的概率也为$\prod_{i=1}^N (1-e_i)$，两者相等。

> Genie-aided译码与SC译码误比特率并不相等，而误块率相等则是因为必须保证所有比特都不出错才代表没有出现误块。
